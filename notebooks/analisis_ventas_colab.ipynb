## 5. Push autenticado con PAT (sin exponer el token)

El token se almacena en **Colab → llave 🔑 → Secrets** con el nombre `GH_PAT` y se recupera vía `userdata.get(...)`. **Nunca** se imprime, ni se pega en una celda, ni queda en el historial de Git. Tras usarlo, la URL remota se restablece para no dejar el PAT residente.

> Si el PAT llegara a filtrarse, debe revocarse de inmediato desde GitHub → Settings → Developer settings → Personal access tokens.

```python
from google.colab import userdata
import subprocess, os

PAT = userdata.get('GH_PAT')          # Token leído desde Secrets de Colab.
USER = 'matiasfnunezdev'
REPO = 'tup-utn-organizacion-empresarial-tp2'
RAMA = 'feature/desarrollo-analisis'

url_segura = f'https://{USER}:{PAT}@github.com/{USER}/{REPO}.git'

# Push con la URL autenticada (un único subprocess, no se persiste en la config).
subprocess.run(['git', 'push', url_segura, RAMA], check=True)

# Limpieza defensiva: la URL inyectada vive solo dentro del comando anterior.
del PAT, url_segura
print('[OK] Push completado sin persistir el token en la configuración Git.')
```
